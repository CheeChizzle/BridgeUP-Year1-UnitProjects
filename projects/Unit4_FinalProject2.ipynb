```
manatee_DNA = open("/content/drive/My Drive/Colab Notebooks/BridgeUP_year1/Unit4_MolecularGenetics/4a_CSN2_DNA.txt").read()
dolphin_DNA = open("/content/drive/My Drive/Colab Notebooks/BridgeUP_year1/Unit4_MolecularGenetics/4b_CSN2_DNA.txt").read()
cow_DNA = open("/content/drive/My Drive/Colab Notebooks/BridgeUP_year1/Unit4_MolecularGenetics/4c_CSN2_DNA.txt").read()
```

```
# translation
conversion = {'A':'U', 'T':'A', 'G':'C', 'C':'G'}
manatee_RNA = ""
dolphin_RNA = ""
cow_RNA = ""
for letter in manatee_DNA:
  manatee_RNA += conversion[letter]

for letter in dolphin_DNA:
  dolphin_RNA += conversion[letter]

for letter in cow_DNA:
  cow_RNA += conversion[letter]
print("Manatee RNA:", manatee_RNA)
print("Dolphin RNA:", dolphin_RNA)
print("Cow RNA:", cow_RNA)
```

```
# Dolphins and Cows 
print("Dolphins and Cows")
differences = 0
similarities = 0
for counter, value in enumerate(dolphin_DNA):
    if value == cow_DNA[counter]:
      similarities +=1
    else:
      differences += 1

percent_diff = 100- (similarities/len(dolphin_DNA) * 100)

print("There are", differences, "differences.")
print("There are", similarities, "similarities.")
print("There is a", round(percent_diff), "percent difference.")
```
