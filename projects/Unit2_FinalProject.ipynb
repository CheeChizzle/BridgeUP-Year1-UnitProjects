```
import matplotlib.pyplot as plt
import numpy as np
import csv # This is a library that allows us to work with CSV files even if we're not using pandas
```

```
csvFile2 = open("/content/drive/My Drive/Colab Notebooks/BridgeUP_year1/Unit2_ClimateChange/sea_temp_anom_Dec_2002.csv", 'r') #'r' means we want to read
csvReader2 = csv.reader(csvFile2, delimiter=',')
data_list2 = list(csvReader2)
```

```
height = len(data_list2)
width = len(data_list2[0])

print(width, height)
```

```
grid = np.empty([height, width, 3], dtype=np.uint8) 
```

```
all_values = [] # We will fill in this empty list with all the data values as we go

for r, row in enumerate(data_list2):
    for c, value in enumerate(row):
        if float(value) < 99998.0:
            all_values.append(float(value)) # This is where we fill in the list with all data values
        if float(value) <= -5:
            grid[r, c] = [255, 255, 255] # white
        elif float(value) <= -4.5:
            grid[r, c] = [0, 0, 0] # black
        elif float(value) <= -4.0:
            grid[r, c] = [80, 0, 100] 
        elif float(value) <= -3.5:
            grid[r, c] = [87, 0, 150] 
        elif float(value) <= -3.0:
            grid[r, c] = [68, 0, 204] 
        elif float(value) <= -2.5:
            grid[r, c] = [25, 0, 255] 
        elif float(value) <= -2.0:
            grid[r, c] = [0, 29, 255] 
        elif float(value) <= -1.5:
            grid[r, c] = [0, 89, 255] 
        elif float(value) <= -1.0:
            grid[r, c] = [0, 152, 255] 
        elif float(value) <= -0.5:
            grid[r, c] = [0, 208, 255]
        elif float(value) <= 0:
            grid[r, c] = [242, 255, 0] 
        elif float(value) <= 0.5:
            grid[r, c] = [255, 238, 0] 
        elif float(value) <= 1.0:
            grid[r, c] = [255, 212, 0] 
        elif float(value) <= 1.5:
            grid[r, c] = [255, 187, 0] 
        elif float(value) <= 2.0:
            grid[r, c] = [255, 161, 0] 
        elif float(value) <= 2.5:
            grid[r, c] = [255, 131, 0]
        elif float(value) <= 3.0:
            grid[r, c] = [255, 102, 0]
        elif float(value) <= 3.5:
            grid[r, c] = [255, 76, 0] 
        elif float(value) <= 4.0:
            grid[r, c] = [255, 51, 0] 
        elif float(value) <= 4.5:
            grid[r, c] = [255, 21, 0] 
        elif float(value) <= 5:
            grid[r, c] = [255, 21, 0]
        else:
            grid[r, c] = [0, 0, 0] # black -- the land
```

```
print(max(all_values)) # max of all values that are NOT 99999
print(min(all_values)) # min of all values that are NOT 99999
```

```
plt.title("Sea Temperature Anomalies, Dec 2002")
plt.imshow(grid, cmap = "jet")
plt.clim(-5, 5)
plt.colorbar(orientation="horizontal",fraction=0.07,anchor=(1.0,0.0))
plt.axis('off')
plt.savefig("Land_temperature.png", dpi = 1000)
# plt.show()
plt.savefig("Sea_Temp_Anom_Dec_2002.png")
```
